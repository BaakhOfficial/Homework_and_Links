# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
# В ЦЕЛОМ ЭТО ВСЕ ВООБЩЕ НЕ НУЖНО
# Не заметил ниже текст и думал что написанное сверху - это одно задание, как оно есть
# Оставляю просто потому что прикольная программа получилась, часа 2 возился, не хочу удалять

def doc_input(keys, **kwargs):
    if kwargs['type'] == 'key':
        doc_inp = input('Какой параметр документа вас интересует?\nВ случае желания выйти введите q')
        if doc_inp in keys: return doc_inp

    elif kwargs['type'] == 'value':
        doc_inp = input('Введите значение данного параметра\nВ случае желания выйти введите q')
        if doc_inp in keys: return doc_inp

    if doc_inp == 'q': return 'q'
    print('Неверное значение, его нет среди наших документов')
    return 'f'

    
def directory_search(key, val, documents, directories):
    # Checking every document
    true_shelves = {}
    for doc in documents:
        # Check if that document is interesting for us
        if doc[key] == val:
            # Searching every shelf for our document
            for shelf, dir_values in directories.items():
                # Checking what we are having on that shelf and have we checked that shelf before
                if doc['number'] in dir_values and shelf not in true_shelves:
                    true_shelves[shelf] = 1
                elif doc['number'] in dir_values and shelf in true_shelves:
                    true_shelves[shelf] += 1
    return true_shelves

def output(true_shelves):
    if true_shelves == {}:
        print('Ваших значений нет в списке')
    else:
        print('Вы можете найти:')
        for shelf, amount in true_shelves.items():
            print(f'{amount} шт. на полке номер {shelf}')

def documents_unboxer(documents, **kwargs):
    output = []

    if kwargs['search'] == 'keys':
        for doc in documents:
            output += doc.keys()
        return set(output)

    elif kwargs['search'] == 'values':
        for doc in documents:
            output += doc.values()
        return set(output)


def document_searcher(documents, directories):
    possible_keys = documents_unboxer(documents, search = 'keys')
    possible_values = documents_unboxer(documents, search = 'values')
    i = 0
    while i < 100:
        i += 1
        search_key = doc_input(possible_keys, type = 'key')
        if search_key == 'q': break
        elif search_key == 'f': continue

        value = doc_input(possible_values, type = 'value')
        if value == 'q': break
        elif value == 'f': continue

        shelves = directory_search(search_key, value, documents, directories)
        output(shelves)

def main():
    documents = [
        {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
        {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
        {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
    ]
    directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
    }
    document_searcher(documents, directories)
    print('Спасибо за использование программы')

main()
        
````

````python
# Подчеркну что условностью тут является то, что номер документа уникален. Иначе это абсурд
# Переписать под иной вариант программу не сложно, но не хочется

# Также я не сделал дополнительные на момент отправки задания, но вернуться в планах потому эти строки закомментированы

def documents_unboxer(documents, type = 'values'):
    output = []
    if type == 'keys':
        for doc in documents:
            output += doc.keys()
        return set(output)
    elif type == 'values':
        for doc in documents:
            output += doc.values()
        return set(output)

def input_check(value, documents, type = 'values'):
    if type == 'keys':
        if value in documents_unboxer(documents, type): return 't'
        elif value == 'q': return 'q'
    elif type == 'values':
        if value in documents_unboxer(documents, type): return 't'
        elif value == 'q': return 'q'
    print('Неверное значение, его нет среди наших документов')
    return 'f'

def number_input(documents):
    doc_no = input('Введите номер документа')
    check = input_check(doc_no, documents)
    if check == 'f': print('Документ не найден в базе')
    return doc_no, check

def owner_number_search(documents):
    check = 0
    while check != 'q':
        doc_no, check = number_input(documents)
        if check == 'q': continue
        for doc in documents:
            if doc['number'] == doc_no:
                print('Владелец документа:', doc['name'])
                break

def shelf_number_search(documents, directories, input = 1, doc_no = 0):
    check = 0
    while check != 'q':
        if input == 1:
            doc_no, check = number_input(documents)
            if check == 'q': continue
        elif input == 0:
            check = 'q'
        for num, shelf in directories.items():
            if doc_no in shelf and input == 1:
                print('Документ хранится на полке:', num)
                break
            elif doc_no in shelf and input == 0:
                return num
                

def doc_info(documents, directories):
    for doc in documents:
        shelf_number_search(documents, directories, input = 0, doc_no = doc['number'])
        print('№', doc['number'], 'тип:', doc['type'], 'полка хранения:', shelf_number_search(documents, directories, input = 0, doc_no = doc['number']))

def dir_unboxer(directories, type = 'keys'):
    output = []
    if type == 'keys':
        for dir in directories.keys():
            output += dir
        return output
    elif type == 'values':
        for dir in directories.values():
            output += dir
        return output

def shelf_edit(cmd = 'ads'):
    global directories
    shelf_num = 0
    while shelf_num != 'q':
        shelf_list = ''
        for x in list(directories.keys()): shelf_list += x + ' '
        print('Текущий перечень полок:', shelf_list)
        shelf_num = input('Введите номер полки')
        if shelf_num == 'q': continue
        elif not shelf_num.isnumeric():
            print('Неверное значение, можно ввести лишь цифру')
            continue
        if cmd == 'ads':
            if shelf_num not in dir_unboxer(directories):
                directories[shelf_num] = []
                print('Полка добавлена.')
            elif shelf_num in dir_unboxer(directories):
                print('Данная полка существует, введите другую.')
        elif cmd == 'ds':
            if shelf_num in dir_unboxer(directories):
                if directories[shelf_num] == []:
                    del directories[shelf_num]
                    print('Полка удалена')
                else:
                    print('В полке присутствуют значения, выберите другую.')
            elif shelf_num in dir_unboxer(directories): print('Данная полка уже существует, введите другую.')

def secretary_helper(documents, directories):
    command = 0
    while command != 'q':
        command = input('Введите команду')
        if command == 'p': owner_number_search(documents)
        elif command == 's': shelf_number_search(documents, directories)
        elif command == 'l': doc_info(documents, directories)
        elif command == 'ads': shelf_edit(cmd = command)
        elif command == 'ds': shelf_edit(cmd = command)
        # elif command == 'ad': add_document()
        # elif command == 'd': delete_document()
        # elif command == 'm': move_document()
        elif command == 'q': continue
        # elif command == 'help': command_info()
        else: print('Команда некорректна, введите повторно\n В любой точке программы вы можете нажать q для перехода на блок выше')

# Main problem here is that it saves data only when we are working in it
# After start it drops all values on default. But I don't know how to work with files right now
# So yeah
def main(documents, directories):
    print('Введите команду')
    secretary_helper(documents, directories)
    print('Спасибо за использование программы')

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

main(documents, directories)
````
