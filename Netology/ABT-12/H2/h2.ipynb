## Продумайте, какие А/Б тесты можно провести, чтобы проверить эти гипотезы.

| Гипотеза | A/B тест |
| - | - |
| Перенапрение рекламного трафика на пользователей Ios | Создание генеральной совокупности из потенциальных клиентов, ранее не использовавших наше приложение с двумя разными видами устройств. В идеале чтобы это была аудитория с лучшей конверсией в лида согласно предыдущей практике. Впоследствии данная аудитория делится на 2 и мы смотрим как они себя ведут в нашем приложении, а также какова стоимость их привлечения |
| Увеличение цены за расстояние | В основном это должно влиять на пользователей бизнес-класса и немного на комфорт, поскольку согласно исследованию (в дз 1 подробно представлено) среднее расстояние проезжаемое ими выше чем у остальных. Следовательно мы берем ту аудиторию, на которую это повлияет сильнее всего - именно их, проводим тестирование и смотрим на первичные метрики, такие как средняя стоимость заказа, а такжже в качестве вторичных метрик используем количество отказов после увиденной цены |
| Рекламная кампания, направленная на людей в возрасте | Поскольку данная рекламная компания никак не затронет основную часть клиентов, нам надо понять заинтересованность наших потенциальных клиентов в нашем продукте. Как следствие мы можем создать совокупность из потенциальных клиентов нашего приложения в возрасте и поделить их на два тем или иным образом, чтобы выборка осталась репрезентативной. Первым мы покажем специализированно направленную на них рекламу, вторым - стандартную рекламу для нашей целевой аудитории. Смотрим на прирост клиентов данной категории, стоимость их привлечения (разницу между этими показателями в первой и второй группе) |
